```python
from nudeform.model import *
from nudeform.view import *
```

```python
A = 72
beta = 0.4
gamma = np.pi/3
granularity = 360
```

```python
phi, theta = np.mgrid[0:2*np.pi:(granularity)*1j, 0:np.pi:(granularity)*1j]
```

```python
Rquadr = quadrupole_shape_scipy(phi,theta, beta = beta, gamma = gamma)
r0 = initial_radius(72)
R = r0*(1+Rquadr)
x,y,z = xyz(R,theta,phi)
```

```python
shape_plot_3D(x, y, z, R, save = False, dpi=300, string = '', axis = False)
plt.show()
```

```python
r_initial = initial_radius(A) * np.ones((360,360))
plot_radius_polar_coordinates(R, r_initial,phi, theta, save = False, dpi=300, string = '')
```

```python
beta = 0.4
for gamma in np.arange(0, np.pi*2, np.pi/6, dtype='g'):
    Rquadr = quadrupole_shape_scipy(phi,theta, beta = beta, gamma = gamma)
    r0 = initial_radius(72)
    R = r0*(1+Rquadr)
    string=f'beta={beta}_gamma={gamma}_A={A}'

    x,y,z = xyz(R,theta,phi)
    shape_plot_3D(x, y, z, R, save = False, dpi=300, string = string, axis = False)
    plt.show()
    plot_radius_polar_coordinates(R, r_initial,phi, theta, save = True, dpi=500, string = string)
```
